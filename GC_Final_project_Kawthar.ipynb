##### One hot encoding

First, we'll change the categorical variables to dummies and add them back to scaled dataset

```python
df_obj = df2.select_dtypes(include='object')
df_num = df2.select_dtypes(include ='float')
df_obj
```

```python
df_obj = pd.get_dummies(df_obj, drop_first=True)
df_obj
```

```python
final_df = pd.concat([df_num, df_obj], axis=1)
final_df
```

```python
#spliting the data
x = final_df.drop(['New_salary'],axis=1)
y= final_df['New_salary']

#I'll scale the data after splitting to make reverse transform easy later 
x = scaler.fit_transform(x)
y = scaler.fit_transform((y).values.reshape(-1,1))

from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split (x, y, test_size = 0.3, random_state = 4)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
#Evaluating the model
from sklearn import metrics
MAD = metrics.mean_absolute_error(y_test, y_pred)
MSE = metrics.mean_squared_error(y_test, y_pred)
print('MAD: ',MAD, '\n', 'MSE: ',MSE)
```

```python
#let's check which of the variables have high imporance by checkng the p value
import statsmodels.api as sm
from scipy import stats
X2 = sm.add_constant(x_train)
est = sm.OLS(y_train, X2)
est2 = est.fit()
print(est2.summary())
```
